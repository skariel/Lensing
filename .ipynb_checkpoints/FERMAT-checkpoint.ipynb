```julia
y'' == 2 Fy/C^2
y'' == -2GM/C^2/r^3 * y
y'' == -K * y

K = 2GM/C^2/r^3 ==> M = K*C^2*r^3/(2G)

K*C^2*r^3/(2G) == 4/3 π r^3 ρ
K == 8/3 G π ρ / C^2

R = 5.0e5
V = 4/3*π*R^3
ρ = 1.0e6/V
K == 8/3*G*π*ρ/C^2

y0 = c1*sin(KS*x0)+c2*cos(KS*x0)
c1 = (y0-c2*c)/s

y'/KS = c1*c-c2*s
c2 = (c1*c-y'/KS)/s
c2 = y0*c/s^2-c2*c^2/s^2-y'/KS/s
c2 = (y0*c/s^2-y'/KS/s)/(1+c^2/s^2)
```

```julia
R = 5.0e5
V = 4/3*π*R^3
ρ = 1.0e7/V
K = 8/3*G*π*ρ/C^2
KS = sqrt(K)

x0 = -5.0e5
y0 = 4.0e5

c = cos(KS*x0)
s = sin(KS*x0)

@show c2 = (y0*c/s^2)/(1+c^2/s^2)
@show c1 = (y0-c2*c)/s;
```

```julia
mfunc(r) = 4/3*π.*r.*r.*r.*ρ
rfunc(r) = ρ
const _x=-5.0e5
const _y=4.0e5
const _rs = sqrt(_x*_x+_y*_y)
const _Θ = abs(asin(_y/_rs));
const _ϕm = π-_Θ
const _tgα = 0.0;

nx, ny, u, du, ia, oa, da, θ =
    numeric_el(mfunc, rfunc, _Θ, _rs, _tgα, 100000; ϕm=_ϕm, reltol=1e-23, abstol=1e-23);

function _phi_a(x,y)
    y += 400000.0
    r = sqrt(x*x+y*y)
    G*mfunc(r)/r
end
function _force_a(x,y)
    y += 400000.0
    const r = sqrt(x*x+y*y)
    const gmr3 = G*mfunc(r)/r/r/r
    gmr3*x, gmr3*y
end
function _force_ay(x,y)
    fx,fy = _force_a(x,y)
    fy
end
fx, fy = fermat(_phi_a, _force_a, 500000.0, _tgα, 100000; abstol=1e-27, reltol=1e-27);
fy+=400000.0;
fx_simple, fy_simple = fermat_y(_force_ay, 500000.0, _tgα, 100000; abstol=1e-27, reltol=1e-27);
fy_simple+=400000.0;


ay = c1*sin(KS*fx)+c2*cos(KS*fx);
ay_simple = c1*sin(KS*fx_simple)+c2*cos(KS*fx_simple);
```

```julia
plot(to_small(nx)/1e3,to_small(ny), label="GR")
hold(true)
plot(to_small(fx)/1e3,to_small(fy),"-y",lw=3.7, label="Fermat, numeric")
plot(to_small(fx)/1e3,to_small(ay),"-r",lw=0.7, label="Fermat, analytic")
plot(to_small(fx_simple)/1e3,to_small(fy_simple),":k",lw=5.7, label="Fermat, simple")

xlabel("X [Mpc/h]")
ylabel("Y [Kpc/h]")
grid(true)
legend(loc=3)
title("testing Fermat solution on LSS scaes")
```

```julia
_x=-1.0e6
_y=4.0e5
_rs = sqrt(_x*_x+_y*_y)
_Θ = abs(asin(_y/_rs));
_ϕm = π-_Θ
_tgα = 0.0;
_M0 = 1.0e7

nx, ny, u, du, ia, oa, da, θ =
    numeric_el(r->_M0, r->0.0, _Θ, _rs, _tgα, 100000; ϕm=_ϕm, reltol=1e-23, abstol=1e-23);
function _phi_4(x,y)
    y += 400000.0
    r = sqrt(x*x+y*y)
    G*_M0/r
end
function _force_4(x,y)
    y += 400000.0
    const r = sqrt(x*x+y*y)
    const gmr3 = G*_M0/r/r/r
    gmr3*x, gmr3*y
end

fx, fy = fermat(_phi_4, _force_4, 1000000.0, _tgα, 100000; abstol=1e-23, reltol=1e-23);
fy+=400000.0;

hold(false)
plot(to_small(nx)/1e3,to_small(ny),label="full GR")
hold(true)
plot(to_small(fx)/1e3,to_small(fy),"--r", label="Fermat")
hold(false)
grid(true)
legend(loc=1)
xlabel("X [Mpc/h]")
ylabel("Y [Kpc/h]")
title("comparing full GR to Fermat for LSS")
```

```julia
N=100000;
z = 0.5;
mvir = 390.0;
concentration = 40.0;
rvir = RVIR(z, mvir);
rscale = rvir/concentration;

rs = 1.0e6;
b_up = 7.0;
b_down=7.6;
Θ = 0.0;
tgα_up = b_up/rs;
tgα_down = b_down/rs;
ϕm_up = π+atan(tgα_up)/4;

function phi(x,y)
    r = sqrt(x*x+y*y)
    tnfw_potential(mvir, rscale, r, rvir);
end
function force(x,y)
    const r = sqrt(x*x+y*y)
    const gmr3 = G*tnfw_m(mvir, rscale, r, rvir)/r/r/r
    gmr3*x, gmr3*y
end

function phi_lss(x,y)
    y_l1 = y-dy_lens_lss
    r = sqrt(x*x+y_l1*y_l1)
    p1 = tnfw_potential(mvir, rscale, r, rvir);
    y_l2 = y+400000.0
    r = sqrt(x*x+y_l2*y_l2)
    p2 = G*_M0/r
    p1+p2
end
function force_lss(x,y)
    y_l1 = y-dy_lens_lss
    r = sqrt(x*x+y_l1*y_l1)
    gmr3 = G*tnfw_m(mvir, rscale, r, rvir)/r/r/r
    fx1 = gmr3*x
    fy1 = gmr3*y_l1
    y_l2 = y+400000.0
    r = sqrt(x*x+y_l2*y_l2)
    gmr3 = G*_M0/r/r/r
    fx2 = gmr3*x
    fy2 = gmr3*y_l2
    fx1+fx2, fy1+fy2
end

#nx_up, ny_up, u, du, ia, oa, da, θ =
#numeric_tnfw_elVir(z, mvir, rscale, Θ, rs, tgα_up, N; ϕm=ϕm_up);
#nx_down, ny_down, u, du, ia, oa, da, θ =
#numeric_tnfw_elVir(z, mvir, rscale, Θ, rs, tgα_down, N);

tx_up, ty_up, u, tu, ia, oa, da, θ =
thin_tnfwVir(z, mvir, rscale, Θ, rs, tgα_up, N; ext=1.3);
tx_down, ty_down, u, tu, ia, oa, da, θ =
    thin_tnfwVir(z, mvir, rscale, Θ, rs, tgα_down, N);

fx_up, fy_up = fermat(phi, force, rs, tgα_up, N; abstol=1e-15, reltol=1e-15, ext=1.3);
fx_down, fy_down = fermat(phi, force, rs, tgα_down, N; abstol=1e-15, reltol=1e-15, ext=1.3);

lx_up, ly_up = fermat(phi_lss, force_lss, rs, tgα_up, N; abstol=1e-15, reltol=1e-15, ext=1.3);
lx_down, ly_down = fermat(phi_lss, force_lss, rs, -tgα_down, N; abstol=1e-15, reltol=1e-15, ext=1.3);


ny_down = -ny_down
ty_down = -ty_down
fy_down = -fy_down

fy_down += linspace(0,0.001,length(fy_down)).^3
fy_up += linspace(0,0.001,length(fy_up)).^3

ly_up -= linspace(0,0.001,length(ly_up)).^3

#nx_up,ny_up, nx_down,ny_down = extrapolate_to_meeting_same_side(nx_up,ny_up, nx_down,ny_down);
#tx_up,ty_up, tx_down,ty_down = extrapolate_to_meeting_same_side(tx_up,ty_up, tx_down,ty_down);
fx_up,fy_up, fx_down,fy_down = extrapolate_to_meeting_same_side(fx_up,fy_up, fx_down,fy_down);
lx_up,ly_up, lx_down,ly_down = extrapolate_to_meeting_same_side(lx_up,ly_up, lx_down,ly_down);
```

```julia
function running_mean(arr, wl)
    a = zeros(length(arr)-wl)
    for i in div(wl,2):(length(arr)-div(wl,2)-1)
        s = i-div(wl,2)+1
        e = i+div(wl,2)+1
        a[i-div(wl,2)+1] = mean(arr[s:e])
    end
    a
end

function load_gx()
    arr = zeros(Float32, 1024)
    open("/Users/ariel/Downloads/adi_acc_x.data","r") do f
        read!(f, arr)
    end
    arr *= 1024^3*G
    append!(arr, arr)
    append!(arr, arr)
    arr
end
arr = running_mean(load_gx(), 6);
```

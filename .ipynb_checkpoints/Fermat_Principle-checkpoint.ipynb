```julia
function ferma_time(x, y, massfunc)
    dr = zeros(length(x)-1)
    for i in 1:(length(dr)-1)
        _dy = y[i+1]-y[i]
        _dx = x[i+1]-x[i]
        _dr = sqrt(_dx*_dx+_dy*_dy)
        cy = y[i]+0.5*_dy
        cx = x[i]+0.5*_dx
        cr=sqrt(cx*cx+cy*cy)
        phi = G.*massfunc(cr)/cr
        dr[i] = _dr #/(1.0-2.*phi/C/C)
    end
    sum(dr)
end

function ferma_grad(x, y, massfunc)
    function dr(x0,y0,x1,y1)
        dx = x1-x0
        dy = y1-y0
        cx = 0.5*(x0+x1)
        cy = 0.5*(y0+y1)
        cr = sqrt(cx*cx+cy*cy)
        phi = G*massfunc(cr)/cr
        _dr = sqrt(dx*dx+dy*dy)
        _dr #/(1.0-2.*phi/C/C)
    end
    gr = zeros(length(x)-2)
    for i in 2:(length(x)-1)
        delta_y = 1.0e-9
        new_y = y[i]+delta_y
        dr_ip = dr(x[i-1],y[i-1],x[i]  ,y[i])
        dr_in = dr(x[i]  ,y[i]  ,x[i+1],y[i+1])
        dr_fp = dr(x[i-1],y[i-1],x[i]  ,new_y)
        dr_fn = dr(x[i]  ,new_y ,x[i+1],y[i+1])
        dt_i = dr_ip+dr_in
        dt_f = dr_fp+dr_fn        
        gr1 = (dt_f-dt_i)./delta_y
        
        delta_y = -1.0e-9
        new_y = y[i]+delta_y
        dr_ip = dr(x[i-1],y[i-1],x[i]  ,y[i])
        dr_in = dr(x[i]  ,y[i]  ,x[i+1],y[i+1])
        dr_fp = dr(x[i-1],y[i-1],x[i]  ,new_y)
        dr_fn = dr(x[i]  ,new_y ,x[i+1],y[i+1])
        dt_i = dr_ip+dr_in
        dt_f = dr_fp+dr_fn        
        gr2 = (dt_f-dt_i)./delta_y

        gr[i-1] = 0.5*(gr1+gr2)
    end
    gr./sum(sqrt(gr.*gr))
end
```

```julia
function ferma_opt(x,y,massfunc)
    it = ferma_time(x,y,massfunc)
    g = ferma_grad(x,y,massfunc)
    y=deepcopy(y)
    y[2:(end-1)] -= g*0.03
    ft = ferma_time(x,y,massfunc)
    x,y
end
```

```julia
z = 0.5;
mvir = 390.0;
concentration = 40.0;
rvir = RVIR(z, mvir);
rscale = rvir/concentration;

rs = 5.0e5;
b = 3.0;
Θ = 0.0;
tgα = b/rs;

nx, ny, u, du, ia, oa, da, θ =
    numeric_tnfw_elVir(z, mvir, rscale, Θ, rs, tgα, 100000);
tx, ty, u, du, ia, oa, da, θ =
    thin_tnfwVir(z, mvir, rscale, Θ, rs, tgα, 100000);
```

```julia
function solve(phifunc, massfunc, rs, tgα, N=100)
    function dphidx(x,y)
       r = sqrt(x*x+y*y)
       massfunc(r)/r/r/r*x*G/C/C        
    end
    function dphidy(x,y)
       r = sqrt(x*x+y*y)
       massfunc(r)/r/r/r*y*G/C/C        
    end
    function phixy(x,y)
        phifunc(sqrt(x*x+y*y))
    end
    function f(x,yvec)
        (y, v) = yvec
        y_prime = v
        v_prime = 0.0
        v_prime = (2*dphidx(x,y)*(v*v*v+v)-2*(v*v+1)^2*dphidy(x,y)) / (1-2*phixy(x,y))
        [y_prime, v_prime]
    end
    _x = linspace(0.0,1,div(N,2)).^2
    _l = [-reverse(_x)*rs ; _x*rs]
    x, u = ode45(f, [0.0; tgα], _l; reltol=1e-13, abstol=1e-13)
    y = map(u -> u[1], u);
    dy = map(u -> u[2], u);    
    x,y
end
function solve_ls(phifunc, massfunc, rs, tgα, N=100)
    const dd=0.00001
    const d=1+dd
    function dphidx(x,y)
       r = sqrt(x*x+y*y)
       p1 = massfunc(r)/r/r/r*x*G/C/C
       
       r = sqrt(x*x+(y+400000.0)*(400000.0))
        p2 = 1.0e6/r/r/r*x*G/C/C
        
       p1+p2 
    end
    function dphidy(x,y)
       r = sqrt(x*x+y*y)
       p1 = massfunc(r)/r/r/r*y*G/C/C        
        r = sqrt(x*x+(y+400000.0)*(400000.0))
        p2 = 1.0e6/r/r/r*(y+400000.0)*G/C/C

        p1+p2
    end
    function phixy(x,y)
        r = sqrt(x*x+(y+400000.0)*(400000.0))
        phifunc(sqrt(x*x+y*y)) + 1.0e6*G/C/C/r # XXXXX
    end
    function f(x,yvec)
        (y, v) = yvec
        y_prime = v
        v_prime = 0.0
        v_prime = (2*dphidx(x,y)*(v*v*v+v)-2*(v*v+1)^2*dphidy(x,y)) / (1-2*phixy(x,y))
        [y_prime, v_prime]
    end
    _x = linspace(0.0,1,div(N,2)).^2
    _l = [-reverse(_x)*rs ; _x*rs]
    x, u = ode45(f, [0.0; tgα], _l; reltol=1e-13, abstol=1e-13)
    y = map(u -> u[1], u);
    dy = map(u -> u[2], u);    
    x,y
end
```

```julia
phifunc(r) = tnfw_potential(mvir, rscale, r, rvir)/C/C;
massfunc(r) = tnfw_m(mvir, rscale, r, rvir);
sux, suy = solve(phifunc, massfunc, rs, tgα, 500000);
sdx, sdy = solve(phifunc, massfunc, rs, -tgα, 500000);

lux, luy = solve_ls(phifunc, massfunc, rs, tgα, 500000);
ldx, ldy = solve_ls(phifunc, massfunc, rs, -tgα, 500000);
```

```julia
plot(to_small(sux)/1e3,to_small(suy[1:length(sux)]),"y-",ms=0.005,lw=3.3, label="Fermat (no LSS)")
hold(true)
plot(to_small(sdx)/1e3,to_small(sdy[1:length(sdx)]),"y-",ms=0.005,lw=3.3)
plot(to_small(nx)/1e3,to_small(ny),"c",label="GR (no LSS)")

plot(tx/1e3,ty,":k",label="thin (no LSS)")
plot(to_small(lux)/1e3,to_small(luy),"--c",lw=3.3,label="Fermat, LSS")
plot(ldx/1e3,ldy,"--c",lw=3.3)

xlim(-750,600)
ylim(-4,4)
xlabel("X [Mpc/h]")
ylabel("Y [Kpc/h]")
grid(true)
legend(loc=2)
text(-700,-3,"\$\\Delta t\$ no LSS: 0 hrs\n\$\\Delta t\$ with LSS: 100 hrs")
```

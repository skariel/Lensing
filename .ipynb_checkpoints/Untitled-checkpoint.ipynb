```julia
z = 0.5;
mvir = 390.0;
concentration = 40.0;
rvir = RVIR(z, mvir);
rscale = rvir/concentration;

rs = 5.0e5;
b = 3.0;
Θ = 0.0;
tgα = b/rs;

nx, ny, u, du, ia, oa, da, θ =
    numeric_tnfw_elVir(z, mvir, rscale, Θ, rs, tgα, 100000);
tx, ty, u, du, ia, oa, da, θ =
    thin_tnfwVir(z, mvir, rscale, Θ, rs, tgα, 100000);
```

```julia
function solve(phifunc, rs, tgα, N=100)
    const dd=0.001
    const d=1+dd
    function dphidx(x,y)
        if x==0.0
            return 0.0
        end
        nx = d*x
        dx = nx-x
        phi1 = phifunc(sqrt(x*x+y*y))
        phi2 = phifunc(sqrt(nx*nx+y*y))
        (phi2-phi1)/dx
    end
    function dphidy(x,y)
        if y==0.0
            return 0.0
        end
        ny = d*y
        dy = ny-y
        phi1 = phifunc(sqrt(x*x+y*y))
        phi2 = phifunc(sqrt(x*x+ny*ny))
        (phi2-phi1)/dy
    end
    function phixy(x,y)
        phifunc(sqrt(x*x+y*y))
    end
    function f(x,yvec)
        (y, v) = yvec
        y_prime = v
        v_prime = 0.0
        v_prime = (2*dphidx(x,y)*(v*v*v+v)-2*(v*v+1)^2*dphidy(x,y)) / (1-2*phixy(x,y))
        [y_prime, v_prime]
    end
    _x = linspace(0.0,1,div(N,2)).^2
    _l = [-reverse(_x)*rs ; _x*rs]
    x, u = ode78(f, [0.0; tgα], _l; reltol=1e-19, abstol=1e-19)
    y = map(u -> u[1], u);
    dy = map(u -> u[2], u);    
    x,y
end
```

```julia
phifunc_3(r) = tnfw_potential(mvir, rscale, r, rvir)/C/C;
sx, sy = solve(phifunc_3, rs, tgα, 100000);
```

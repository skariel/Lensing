```julia
include("cosmology.jl")
using PyPlot, ODE
set_bigfloat_precision(128);
```

```julia
function _sigrp(rp, R)
    c1 = sqrt(R*R./rp./rp-1.0)
    1./rp.*atan(c1)
end
function _dmdrp(rp, R)
    rp.*_sigrp(rp,R)
end
function _mrp(rp, R)
    quadgk(rp->_dmdrp(rp, R), 0.0, rp)[1]
end
function mrp(M0, R, rp)
    f = M0/_mrp(R, R)
    _mrp(rp, R)*f
end
```

```julia
rg(m) = 2.0.*m.*G./C./C
function miso(M0, R, r)
    M0.*r./R
end
rgiso(M0, R, r) = rg(miso(M0, R, r))

function direct_lens(M0, R, b, N=100)
    # only inside the halo
    
    ω=sqrt(1.0-3M0*G/C/C/R)
    tgα = b/R
    dudθ = 1.0/R/tgα
    u1 = 1.0/R
    ϕ = atan(-dudθ/ω/u1)    
    
    ϕm=π/2-ω*ϕ*1.000002
    ϕl = linspace(0,ϕm,N) 
    function f(ϕ, y)
        (u, v) = y
        u_prime = v
        v_prime = 3*M0*G/C/C/R*u-u
        [u_prime; v_prime]    
    end
    const u0 = 1.0/R   
    const du0 = dudθ
    start = [u0; du0];
    
    ϕ, y = ode45(f, start, ϕl, abstol=1.0e-13, reltol=1.0e-14);
    u = map(y -> y[1], y);
    du = map(y -> y[2], y);

    x = [Float64(v) for v in -cos(ϕ)./u]
    y = [Float64(v) for v in sin(ϕ)./u]
    x,y, u[end], du[end]
end

function direct_lens3(M0, R, b, u0, du0, N=100)
    # hitting the halo but defined outside too
    ϕl = linspace(0, π, N)    
    function f(ϕ, y)
        (u, v) = y
        u_prime = v
        v_prime = 3*M0*G/C/C/R*u-u
        [u_prime; v_prime]    
    end
    start = [u0; du0];
    
    ϕ, y = ode45(f, start, ϕl, abstol=1.0e-13, reltol=1.0e-14);
    u = map(y -> y[1], y);
    
    x = [Float64(v) for v in -cos(ϕ)./u]
    y = [Float64(v) for v in sin(ϕ)./u]        
    x,y
end

function direct_lens5(M0, R, b, N=100)
    # mising the halo
    rg = M0*G/C/C*2
    function f(ϕ, y)
        (u, v) = y
        u_prime = v
        v_prime = 3/2*rg*u*u-u
        [u_prime; v_prime]    
    end
    const u0 = 1.0/R
    const tgθ0 = b/R
    const du0 = 1.0/(R*abs(tgθ0))
    start = [u0; du0];
    ϕl = linspace(0, π, N)        
    ϕ, y = ode45(f, start, ϕl, abstol=1.0e-13, reltol=1.0e-14);
    u = map(y -> y[1], y);
    x = [Float64(v) for v in -cos(ϕ)./u]
    y = [Float64(v) for v in sin(ϕ)./u]        
    x,y
end

function analytic_lens(M0, R, b, N=100)
    ω=sqrt(1.0-3M0*G/C/C/R)
    
    tgα = b/R
    dudθ = -1.0/R/tgα
    u1 = 1.0/R
    
    ϕ = atan(dudθ/ω/u1)
    A = u1/cos(ϕ)
    
    ϕm=π/2-ω*ϕ*1.000002
    
    ϕl = linspace(0.0,ϕm,N)
    u = A*cos(ω*ϕl+ϕ)
    ixs = find(u.>0)
    u=u[ixs]
    ϕl = ϕl[ixs]
    
    x = [Float64(v) for v in -cos(ϕl)./u]
    y = [Float64(v) for v in sin(ϕl)./u]        
    x,y
end

function thin_lens(M0p, R, b, N=100)
    const tgθ = b/R
    const θ=atan(tgθ)
    #const nθ = θ-2.0*   frg(b)   /b
    const nθ = θ-4*M0p*G/C/C   /b
    const tgnθ = tan(nθ)
    nx = -b/tgnθ
    if nθ>0.0
        nx = R
    end
    
    lx = linspace(0.0,1.0,div(N,2)).^3
    x = [-reverse(lx)*R; lx*nx]
    y = zeros(length(x))
    for i in 1:length(x)
        y[i] = if x[i]<0.0
            tgθ*(x[i]+R)
        else
            b+tgnθ*x[i]
        end
    end
    x, y
end

function thin_lens_big(M0p, R, b, N=100)
    M0p = big(M0p)
    R=big(R)
    b=big(b)
    
    # testing with bigfloats
    const tgθ = b/R
    const θ=atan(tgθ)
    #const nθ = θ-2.0*   frg(b)   /b
    const nθ = θ-4*M0p*G/C/C   /b
    const tgnθ = tan(nθ)
    nx = -b/tgnθ
    if nθ>0.0
        nx = R
    end
    
    lx = [big(i) for i in linspace(0.0,1.0,div(N,2)).^3]
    x = [big(i) for i in [-reverse(lx)*R; lx*nx]]
    y = [big(i) for i in zeros(length(x))]
    for i in 1:length(x)
        y[i] = if x[i]<0.0
            tgθ*(x[i]+R)
        else
            b+tgnθ*x[i]
        end
    end
    [Float64(i) for i in x], [Float64(i) for i in y]
end

function thin_lens2(M0p, R, b, N=100)
    # just calculating the deflection angle
    const tgθ = b/R
    const θ=atan(tgθ)
    #const nθ = θ-2.0*   frg(b)   /b
    const deflection_angle = 4*M0p*G/C/C   /b
    const nθ = θ-deflection_angle
    const tgnθ = tan(nθ)
    nx = -b/tgnθ
    if nθ>0.0
        nx = R
    end
    
    lx = linspace(0.0,1.0,div(N,2)).^3
    x = [-reverse(lx)*R; lx*nx]
    y = zeros(length(x))
    for i in 1:length(x)
        y[i] = if x[i]<0.0
            tgθ*(x[i]+R)
        else
            b+tgnθ*x[i]
        end
    end
    x, y, deflection_angle
end


function mfac(M0, R, rp, N, minr=1.0e-3)
    tot = 0.0
    ptot = 0.0
    @inbounds for i in 1:N
        x=rand()*rp+minr
        y=rand()*rp+minr
        z=rand()*R+minr
        r=sqrt(x*x+y*y+z*z)
        r>R && continue
        pr = sqrt(x*x+y*y)
        pr > rp && continue
        ρ=1/r/r
        ptot += ρ
        if r <= rp
            tot += ρ
        end
    end
    ptot/tot
end
```

```julia
Mh = 3.3e6;
Rh = 3.3e6;
b= 4.0;
```

```julia
tas = Float64[]
aas = Float64[]
das = Float64[]
bs = Float64[]
for b in linspace(1.0,12.0,15)
    push!(bs, b)
    
    ai = atan(b/Rh)
    ax, ay = analytic_lens(Mh, Rh, b, 1000)
    dy = ay[end]-ay[end-1]
    dx = ax[end]-ax[end-1]
    af = atan(dy/dx)
    push!(aas, af-ai)


    M0p = mrp(Mh, Rh, b)
    tx,ty,da = thin_lens2(M0p, Rh, b, 1000);
    push!(tas, da)
    
    ai = atan(b/Rh)
    dx, dy, u, du = direct_lens(Mh, Rh, b, 1000)    
    ady = ay[end]-ay[end-1]
    adx = ax[end]-ax[end-1]
    af = atan(ady/adx)
    push!(das, af-ai)


    
    plot(dx/1e3,dy,"-y",lw=3.3)
    hold(true)
    plot(tx/1e3,ty,"-r",lw=0.9)
    plot(ax/1e3,ay,"-k",lw=1)
end
xlim(-Rh/1e3,Rh/1e3)
ylim(0,12)
xlabel("X [Mpc/h]")
ylabel("Y [Kpc/h]")
grid(true)
legend(["Numerical","Thin","Txact"],loc=2)
title("Infinite Isothermal Halo")
```

```julia
plot(bs, tas, "-r")
hold(true)
plot(bs, -das, "-y",lw=3.3)
plot(bs, -aas, "-k")
grid(true)
xlabel("impact parameter b [Kpc/h]")
ylabel("deflection angle")
legend(["Thin","Numerical","Exact"], loc=6)
title("Deflection angles vs impact for infinite isothermal halo")
```

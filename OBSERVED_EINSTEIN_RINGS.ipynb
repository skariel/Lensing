# A370

```julia
z_lens = 0.375
z_source = 1.5
Einstein_radius_as = 43.0
Mvir = 2.93e5
concentration = 7.75

rvir = RVIR(z_lens, Mvir)
r_scale = rvir / concentration

b = Da_as(z_lens, Einstein_radius_as)
distance_lens_source = Dc(z_source)-Dc(z_lens)
distance_lens_observer = Dc(z_lens)

mass(r) = lnfw_m(Mvir, r_scale, r, rvir);
pmass(r) = lnfw_mp(Mvir, r_scale, r, rvir);
rho(r) = lnfw_ρ(Mvir, r_scale, r, rvir);

new_concentration = concentration / 1.52
new_mvir = Mvir

new_rvir = RVIR(z_lens, new_mvir)
new_r_scale = new_rvir / new_concentration;

new_mass(r) = lnfw_m(new_mvir, new_r_scale, r, new_rvir);
new_rho(r) = lnfw_ρ(new_mvir, new_r_scale, r, new_rvir);

tg = b/distance_lens_source

tx, ty, tu, tdu, tia, toa, tda, tθ =
    thin(pmass, 0.0, distance_lens_source, tg, 100000);

ex, ey, eu, edu, eia, eoa, eda, eθ =
    numeric_el(mass, rho, 0.0, distance_lens_source, tg, 100000);

nex, ney, neu, nedu, neia, neoa, neda, neθ =
    numeric_el(new_mass, new_rho, 0.0, distance_lens_source, tg, 100000);

tx, ty = extrapolate_to_x2(tx,ty, distance_lens_observer)
ex, ey = extrapolate_to_x2(ex,ey, distance_lens_observer)
nex, ney = extrapolate_to_x2(nex,ney, distance_lens_observer)


hold(false)
plot(tx/1e3,ty,"-r",lw=0.9, label="thin, c=$(concentration)")
hold(true)
plot(ex/1e3,ey,"-g",lw=1.1, label="num+el, c=$(concentration)")
plot(nex/1e3,ney,"--y",lw=3.1, label="num+el, c=$(round(new_concentration,2))")

hold(false)
xlim(-distance_lens_source*1.1/1e3,distance_lens_observer*1.1/1e3)
ylim(-1.5b, 1.2b)
grid(true)
xlabel("X [Mpc/h]")
ylabel("Y [Kpc/h]")
title("Cluster A370")
legend(loc=3)
text(-2000, -120, "\$M_{vir}=$(Mvir/1e5) \\times 10^{15} M_o/h\$ \n\$Z_{lens}=$(z_lens)\$, \$Z_{source}=$(z_source)\$\n\$\\Theta_{Einstein}=$(Einstein_radius_as)\\ arcsec\$\n\$c_{nfw}=$(concentration)\$, \$R_{vir}=$(round(rvir/1000,3)) Mpc/h\$\n\$b=$(round(Int64,b)) Kpc/h\$")
```
